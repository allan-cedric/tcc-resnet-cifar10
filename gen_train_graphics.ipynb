```python
# Modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
training_log_dfs = {}
for n in [8, 14, 20, 26]:
    training_log_file = f'./models/resnet{n}/training_log.csv'
    training_log_df = pd.read_csv(training_log_file)
    training_log_dfs[str(n)] = training_log_df
training_log_dfs["8"].head(), training_log_dfs["14"].head()
```

```python
for k, df in training_log_dfs.items():
    df["loss"].plot(label=f'ResNet-{k}')

# plt.title("Training losses")
plt.xlabel('poca')
plt.ylabel('Custo (crossentropy loss)')
# plt.yticks(np.arange(0.5, 4, step=0.5))
plt.ylim(0.1, 2)
plt.legend(loc='upper right')
plt.show()
```

```python
for k, df in training_log_dfs.items():
    df["val_loss"].plot(label=f'ResNet-{k}')

# plt.title("Validation losses")
plt.xlabel('poca')
plt.ylabel('Custo (crossentropy loss)')
# plt.yticks(np.arange(0.5, 4, step=0.5))
plt.ylim(0.1, 2)
plt.legend(loc='upper right')
plt.show()
```
