```python
# Modules
import glob 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
power_log_dfs = {}
for n in [8, 14, 20, 26]:
    power_log_file = glob.glob(f'./results/*resnet{n}/*resnet{n}.csv')
    print(power_log_file)
    power_log_df = pd.read_csv(power_log_file[0], delimiter=';')
    power_log_dfs[str(n)] = power_log_df
power_log_dfs["8"], power_log_dfs["14"]
```

```python
power_log_dfs["8"].describe()
```

```python
for k, df in power_log_dfs.items():
    df["Power"][:500].plot(label=f'ResNet-{k}')

plt.title("Power x Time")
plt.xlabel('Time')
plt.ylabel('Power (W)')
plt.yticks(np.arange(0, 1, step=0.1))
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```
