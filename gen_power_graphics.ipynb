```python
# Modules
import glob 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ESP32 Power DataFrames
esp32_power_log_dfs = {}
for n in [8, 14, 20, 26]:
    power_log_file = glob.glob(f'./results/*resnet{n}/*resnet{n}.csv')
    print(power_log_file)
    power_log_df = pd.read_csv(power_log_file[0], delimiter=';')
    esp32_power_log_dfs[str(n)] = power_log_df
esp32_power_log_dfs["8"], esp32_power_log_dfs["14"]
```

```python
# RPi Tflite Power DataFrames
rpi_tflite_power_log_dfs = {}
for n in [8, 14, 20, 26]:
    power_log_file = glob.glob(f'./results/*resnet{n}*tflite-pi/*resnet{n}*.csv')
    print(power_log_file)
    power_log_df = pd.read_csv(power_log_file[0], delimiter=';')
    rpi_tflite_power_log_dfs[str(n)] = power_log_df
rpi_tflite_power_log_dfs["8"], rpi_tflite_power_log_dfs["14"]
```

```python
# RPi Tf Power DataFrames
rpi_tf_power_log_dfs = {}
for n in [8, 14, 20, 26]:
    power_log_file = glob.glob(f'./results/*resnet{n}*tf-pi/*resnet{n}*.csv')
    print(power_log_file)
    power_log_df = pd.read_csv(power_log_file[0], delimiter=';')
    rpi_tf_power_log_dfs[str(n)] = power_log_df
rpi_tf_power_log_dfs["8"], rpi_tf_power_log_dfs["14"]
```

```python
esp32_power_log_dfs["8"].describe(), rpi_tflite_power_log_dfs["8"].describe(), rpi_tf_power_log_dfs["8"].describe()
```

```python
# Exemplo de media e max
for n in [8, 14, 20, 26]:
    print(f"{n}:", esp32_power_log_dfs[str(n)]["Power"].mean(), esp32_power_log_dfs[str(n)]["Power"].max())
```

```python
# Exemplo de grafico de linha sobre potencia
for k, df in esp32_power_log_dfs.items():
    df["Power"][:500].plot(label=f'ResNet-{k}')

plt.title("Power x Time")
plt.xlabel('Time')
plt.ylabel('Power (W)')
plt.yticks(np.arange(0, 1, step=0.1))
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```

```python
resnets = ("ResNet-8", "ResNet-14", "ResNet-20", "ResNet-26")

power_means = {}
power_means['ESP32 (TFLite)'] = [round(esp32_power_log_dfs[str(n)]["Power"].mean(), 3) for n in [8, 14, 20, 26]]
power_means['RPi (TFLite)'] = [round(rpi_tflite_power_log_dfs[str(n)]["Power"].mean(), 3) for n in [8, 14, 20, 26]]
power_means['RPi (TF)'] = [round(rpi_tf_power_log_dfs[str(n)]["Power"].mean(), 3) for n in [8, 14, 20, 26]]

x = np.arange(len(resnets))  # the label locations
width = 0.3  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained', figsize=(8, 6))

for attribute, measurement in power_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Power (W)')
ax.set_title('Power means in all platforms')
ax.set_xticks(x + width, resnets)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 4.5)

plt.show()
```
